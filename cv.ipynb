```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('/content/download.jpeg')
```

```python
c=255/(np.log(1+np.max(img)))
log_transformed = c*np.log(1+img)
```

```python
log_transformed=np.array(log_transformed,dtype=np.uint8)
```

```python
cv2.imwrite('log_transformed.jpg',log_transformed)
plt.imshow(log_transformed)
```

```python
for gama in [0.1,0.5,1.2,2.5]:
  gamma_corrected=np.array(255*(img/255)**gama,dtype='uint8')
  cv2.imwrite('gamma_transformed'+str(gama)+'.jpg',gamma_corrected)
  plt.imshow(gamma_corrected)
```
