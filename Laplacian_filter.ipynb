```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/grayscale.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
#laplacian filter
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

```python
#8 bit format
laplacian_abs = cv2.convertScaleAbs(laplacian)
```

```python
enhanced = cv2.addWeighted(image, 1.5, laplacian_abs, -0.5, 0)
```

```python
cv2_imshow(image)
```

```python
cv2_imshow(laplacian_abs)
```

```python
cv2_imshow(enhanced)
```
